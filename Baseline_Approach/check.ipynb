## Visualize Real-word user movement

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.animation as animation

# Load the CSV
df = pd.read_csv('45minutes_FINAL.csv')
field_size = 30000

# Set up the figure and axis
fig, ax = plt.subplots(figsize=(7, 7))
ax.set_xlim(0, field_size)
ax.set_ylim(0, field_size)
ax.set_title('User Movement Animation')
ax.set_xlabel('X')
ax.set_ylabel('Y')
scat = ax.scatter([], [], s=10, c='blue', alpha=0.7)

def update(frame):
    ax.clear()
    ax.set_xlim(0, field_size)
    ax.set_ylim(0, field_size)
    ax.set_title(f'User Movement at Minute {frame}')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    data = df[df['step'] == frame]
    ax.scatter(data['x'], data['y'], s=10, c='blue', alpha=0.7)
    return scat,

ani = animation.FuncAnimation(fig, update, frames=range(df['step'].min(), df['step'].max()+1), interval=200, blit=False)
# Save GIF files 
ani.save('100_user_movement.gif', writer='pillow', fps=5)
plt.show()
```
