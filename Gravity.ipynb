# Gravity-Based UAV Addition Model: A Visualization Example

#### A new UAV is deployed in the operational area.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Circle

# Parameters

FIELD = 15000

component1 = [
    (3200, 3000),   
    (4200, 4200),   
    (3600, 6000),   
    (3000, 8200),   
    (5200, 8200)    
]

links1 = [
    (0, 1), (1, 2), (2, 0),     
    (1, 2), (2, 3), (1, 3),     
    (3, 4), (2, 4)              
]

component2 = [
    (10000, 8200),
    (11500, 9000),
    (10800, 12500)
]

links2 = [(0, 1), (1, 2), (0, 2)]

# Exploration UAV

uav_start = (6000, 1000)
fig, ax = plt.subplots(figsize=(7.5, 7.5))
ax.set_xlim(0, FIELD)
ax.set_ylim(0, FIELD)
ax.set_aspect("equal")
ax.grid(True, linestyle="--", alpha=0.3)
ax.set_xlabel("X-Coordinate (meters)", fontsize=25)
ax.set_ylabel("Y-Coordinate (meters)", fontsize=25)
ax.tick_params(axis="both", labelsize=20)
plt.xticks(rotation=45)

for i, j in links1:
    ax.plot(
        [component1[i][0], component1[j][0]],
        [component1[i][1], component1[j][1]],
        "k-", lw=2
    )
for idx, p in enumerate(component1, start=1):
    ax.scatter(*p, s=100, color="#1f77b4",
               edgecolors="black", zorder=5)
    ax.text(p[0], p[1], str(idx),
            color="white", fontsize=12,
            ha="center", va="center", fontweight="bold")

for i, j in links2:
    ax.plot(
        [component2[i][0], component2[j][0]],
        [component2[i][1], component2[j][1]],
        "k-", lw=2
    )

for p in component2:
    ax.scatter(*p, s=100, color="#d62728",
               edgecolors="black", zorder=5)
    
ticks = np.arange(0, 15001, 3000)

ax.set_xticks(ticks)
ax.set_yticks(ticks)

ax.scatter(*uav_start, s=140, marker="s",
           color="#f39c12", edgecolors="black", zorder=6)

# Hop 
for r in [4000, 8000, 12000]:
    ax.add_patch(
        Circle(uav_start, r, fill=False,
               linestyle=":", lw=2,
               edgecolor="green", alpha=0.6)
    )

ax.plot(
    [uav_start[0], component1[1][0]],
    [uav_start[1], component1[1][1]],
    linestyle="--", lw=2.5, color="#1f77b4"
)

ax.plot(
    [uav_start[0], component2[0][0]],
    [uav_start[1], component2[0][1]],
    linestyle="--", lw=2.5, color="#d62728"
)

plt.tight_layout()
plt.show()
```

#### UAV attached with the component having the highest gravitational pull.

```python
import numpy as np
import matplotlib.pyplot as plt


# Parameters

FIELD = 15000

component1 = [
    (3200, 3000),
    (4200, 4200),
    (3600, 6000),
    (3000, 8200),
    (5200, 8200)
]

links1 = [
    (0, 1), (1, 2), (2, 0),
    (1, 2), (2, 3), (1, 3),
    (3, 4), (2, 4)
]

component2 = [
    (10000, 8200),
    (11500, 9000),
    (10800, 12500)
]

links2 = [(0, 1), (1, 2), (0, 2)]

uav_final = (4300, 5200)  

fig, ax = plt.subplots(figsize=(7.5, 7.5))
ax.set_xlim(0, FIELD)
ax.set_ylim(0, FIELD)
ax.set_aspect("equal")
ax.grid(True, linestyle="--", alpha=0.3)
ax.set_xlabel("X-Coordinate (meters)", fontsize=25)
ax.set_ylabel("Y-Coordinate (meters)", fontsize=25)
ax.tick_params(axis="both", labelsize=25)
plt.xticks(rotation=45)
ticks = np.arange(0, 15001, 3000)
ax.set_xticks(ticks)
ax.set_yticks(ticks)

for i, j in links1:
    ax.plot(
        [component1[i][0], component1[j][0]],
        [component1[i][1], component1[j][1]],
        "k-", lw=2
    )
for p in component1:
    ax.scatter(*p, s=100, color="#1f77b4",
               edgecolors="black", zorder=5)

for i, j in links2:
    ax.plot(
        [component2[i][0], component2[j][0]],
        [component2[i][1], component2[j][1]],
        "k-", lw=2
    )

for p in component2:
    ax.scatter(*p, s=100, color="#d62728",
               edgecolors="black", zorder=6)

ax.scatter(*uav_final, s=160, marker="s",
           color="#f39c12", edgecolors="black", zorder=6)

ax.plot(
    [uav_final[0], component1[2][0]],
    [uav_final[1], component1[2][1]],
    linestyle="-", lw=3, color="#1f77b4"
)

plt.tight_layout()
plt.show()
```
