# 100 Users and 30 X 30 sq. km. filed

```python
import pandas as pd
import numpy as np
import random

# Parameters
num_users = 100
field_size = 30000  
steps = list(range(75))  
cluster_size = 15000  
margin = 30  

# Region bounds

def get_region(step):
    if step < 15:
        # Left bottom
        x_range = (margin, cluster_size)
        y_range = (margin, cluster_size)
    elif step < 30:
        # Right top
        x_range = (field_size - cluster_size, field_size - margin)
        y_range = (field_size - cluster_size, field_size - margin)
    elif step < 45:
        # Left top
        x_range = (margin, cluster_size)
        y_range = (field_size - cluster_size, field_size - margin)
    elif step < 60:
        # Right bottom
        x_range = (field_size - cluster_size, field_size - margin)
        y_range = (margin, cluster_size)
    else:
        # Middle area
        mid = field_size // 2
        half = cluster_size // 2
        x_range = (mid - half, mid + half)
        y_range = (mid - half, mid + half)
    return x_range, y_range

# Data Generation

data = []
for uid in range(num_users):
    for step in steps:
        x_range, y_range = get_region(step)
        x = random.uniform(*x_range)
        y = random.uniform(*y_range)
        sim_time_min = step
        data.append({'uid': uid, 'step': step, 'x': x, 'y': y, 'sim_time_min': sim_time_min})

# Create CSV file

df = pd.DataFrame(data)
df.to_csv('Dataset/Coverage_recover_and_degradation_analysis_100.csv', index=False)
print("Test CSV generated with", len(df), "rows.")
```

#### sav gif file

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.animation as animation

# Load the CSV
df = pd.read_csv('Dataset/Coverage_recover_and_degradation_analysis_100.csv')
field_size = 30000

# Set up the figure and axis
fig, ax = plt.subplots(figsize=(7, 7))
ax.set_xlim(0, field_size)
ax.set_ylim(0, field_size)
ax.set_title('User Movement Animation')
ax.set_xlabel('X')
ax.set_ylabel('Y')
scat = ax.scatter([], [], s=10, c='blue', alpha=0.7)

def update(frame):
    ax.clear()
    ax.set_xlim(0, field_size)
    ax.set_ylim(0, field_size)
    ax.set_title(f'User Movement at Minute {frame}')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    data = df[df['step'] == frame]
    ax.scatter(data['x'], data['y'], s=10, c='blue', alpha=0.7)
    return scat,

ani = animation.FuncAnimation(fig, update, frames=range(df['step'].min(), df['step'].max()+1), interval=200, blit=False)
# Save GIF files 
ani.save('100_user_movement.gif', writer='pillow', fps=5)
plt.show()
```

## 40 X 40 sq. km. field with 200 users

```python
import pandas as pd
import numpy as np
import random

# Parameters
num_users = 200
field_size = 40000      
cluster_size = 15000    
steps = list(range(75)) 
margin = 30             

# Region bounds
def get_region(step):

    if step < 15:
        # Left bottom
        x_min = margin
        x_max = cluster_size
        y_min = margin
        y_max = cluster_size
    elif step < 30:
        # Right top
        x_min = field_size - cluster_size
        x_max = field_size - margin
        y_min = field_size - cluster_size
        y_max = field_size - margin
    elif step < 45:
        # Left top
        x_min = margin
        x_max = cluster_size
        y_min = field_size - cluster_size
        y_max = field_size - margin
    elif step < 60:
        # Middle area
        x_min = field_size - cluster_size
        x_max = field_size - margin
        y_min = margin
        y_max = cluster_size
    else:
        # Middle area
        center = field_size / 2
        half_cluster = cluster_size / 2
        x_min = center - half_cluster
        x_max = center + half_cluster
        y_min = center - half_cluster
        y_max = center + half_cluster
    return (x_min, x_max), (y_min, y_max)

# Data Generation
data = []
for uid in range(num_users):
    for step in steps:
        x_range, y_range = get_region(step)
        x = random.uniform(*x_range)
        y = random.uniform(*y_range)
        data.append({
            'uid': uid,
            'step': step,
            'x': x,
            'y': y,
            'sim_time_min': step
        })

# Create CSV file
df = pd.DataFrame(data)
df.to_csv('Dataset/Coverage_recover_and_degradation_analysis_40km.csv', index=False)
print("Test CSV generated with", len(df), "rows.")
```
